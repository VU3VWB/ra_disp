```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vec_length = 4096

sinc_sample_locations = np.arange(-np.pi*4/2.0, np.pi*4/2.0, np.pi/vec_length)
sinc = np.sinc(sinc_sample_locations/np.pi)
custom_window = sinc*np.hamming(4*vec_length)

t = np.linspace(0,12,vec_length*4)
inp_data = np.random.normal(0,0.001,len(t)) + np.sin(100.12547*2*np.pi*t)

sig_1 = inp_data[0*vec_length:1*vec_length]*custom_window[0*vec_length:1*vec_length]
sig_2 = inp_data[1*vec_length:2*vec_length]*custom_window[1*vec_length:2*vec_length]
sig_3 = inp_data[2*vec_length:3*vec_length]*custom_window[2*vec_length:3*vec_length]
sig_4 = inp_data[3*vec_length:4*vec_length]*custom_window[3*vec_length:4*vec_length]

sig_full = sig_1+sig_2+sig_3+sig_4

plt.figure(figsize=(16,9))
plt.semilogy(np.abs(np.fft.rfft(sig_full)))
plt.semilogy(np.abs(np.fft.rfft(inp_data[0:vec_length])))
```
